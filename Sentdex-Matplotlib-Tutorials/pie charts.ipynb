```python
import matplotlib.pyplot as plt
```

```python
values = [50, 10, 25, 15]
value_names = ['Gold', 'Silver', 'Titanium', 'Lead']
color_list = ['c', 'm', 'r', 'b']
```

```python
plt.pie(values,
       labels = value_names,
       colors = color_list,
       startangle=180,
       shadow =True,
       explode=(0.3,0.1,0.1,0),
       autopct= '%1.1f%%')

plt.title('Metal Reserve')
```
