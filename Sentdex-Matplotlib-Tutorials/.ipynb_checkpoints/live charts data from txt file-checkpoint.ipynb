```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style
```

```python
style.use('fivethirtyeight')
```

```python
#creating the function for animation
def animate(i):
    graphd = open('C:\\Users\\Admin\\Desktop\\data sets\\ex_data.txt').read()
    lines = graphd.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line) > 1:
            x,y = line.split(',')
            xs.append(x)
            ys.append(y)
    ax1.clear()
    ax1.plot(xs, ys)
            
```

```python
#calling the animation function from matplotlib
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ani = animation.FuncAnimation(fig, animate, interval=1000)
plt.show()
```
