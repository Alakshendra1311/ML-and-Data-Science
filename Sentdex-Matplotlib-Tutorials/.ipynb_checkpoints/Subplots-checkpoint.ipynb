```python
import random
import matplotlib.pyplot as plt
```

```python
#create function to generate random values
def random_val():
    xs = []
    ys = []
    
    for i in range(10):
        x = i
        y = random.randrange(10)
        
        xs.append(x)
        ys.append(y)
    return xs,ys
```

# by using add_subplot function

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

x,y = random_val()
ax1.plot(x,y)

x,y = random_val()
ax2.plot(x,y)
```

```python
fig = plt.figure()
ax3 = fig.add_subplot(221)
ax4 = fig.add_subplot(222)
ax5 = fig.add_subplot(212)


x,y = random_val()
ax3.plot(x,y)

x,y = random_val()
ax4.plot(x,y)

x,y = random_val()
ax5.plot(x,y)
```

```python
fig = plt.figure()
ax6 = fig.add_subplot(221)
ax7 = fig.add_subplot(222)
ax8 = fig.add_subplot(223)
ax9 = fig.add_subplot(224)

x,y = random_val()
ax6.plot(x,y)

x,y = random_val()
ax7.plot(x,y)

x,y = random_val()
ax8.plot(x,y)

x,y = random_val()
ax9.plot(x,y)
```

# by using grid method a.k.a subplot2grid() function

```python
ax10 = plt.subplot2grid((6,1), (0,0), rowspan=2, colspan=1)
ax11 = plt.subplot2grid((6,1), (2,0), rowspan=2, colspan=1)
ax12 = plt.subplot2grid((6,1), (4,0), rowspan=2, colspan=1)

x, y = random_val()
ax10.plot(x,y)

x, y = random_val()
ax11.plot(x,y)

x,y =random_val()
ax12.plot(x,y)
```
