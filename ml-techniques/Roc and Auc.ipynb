```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#to preprocess the data using StandardScaler =>
import sklearn.preprocessing as prep
#to create a confusion matrix =>
from sklearn.metrics import confusion_matrix
```

```python
#standardscalar = prep.StandardScaler
#load data in 'dt' from csv file
dt = pd.read_csv("C:\\Users\\Admin\\Desktop\\datasets\\Social_Network_Ads.csv")
dt.head(10)
```

```python
#real_x is a 2D array which contans age and estimated salary in each of it's cell 
real_x = dt.iloc[:, [2,3]].values
real_x
```

```python
#real_y is a 1D array which contains the purchased value either 1 or 0
real_y = dt.iloc[:, 4].values
real_y
```

```python
#train_test_split function for real_x and real_y
train_x, test_x, train_y, test_y = train_test_split(real_x, real_y, test_size=0.25, random_state=0)
```

```python
train_x
```

```python
#Standard scaler reduces the gap between age and salary it creates a floating
#numbers 

scaler = prep.StandardScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.fit_transform(test_x)
train_x
```

```python
#create classifier object for LogisticRegression class
classifier_LR = LogisticRegression(random_state = 0)
#train the algorithm
classifier_LR.fit(train_x, train_y)
```

```python
#give the test_x and take the prediction 
y_pred = classifier_LR.predict(test_x)
y_pred
```

```python
c_m = confusion_matrix(test_y, y_pred)
c_m
```
