## Data Cleansing (convert categorical data)

```python
# first make a copy
bdt = bank.copy()
```

```python
# job post by deposit 
un_values_job = bdt.job.unique()

for j in un_values_job:
    print('{:15} : {:5}'.format(j,\
                                len(bdt[ (bdt.deposit == 'yes') & (bdt.job == j)])))
```

```python
# Different types of job categories and their counts
bdt.job.value_counts()
```

```python
# combine similar jobs into categories

bdt['job'] = bdt['job'].replace(['management', 'admin.'], 'white-collar')

bdt['job'] = bdt['job'].replace(['services', 'housemaid'], 'pink-collar')

bdt['job'] = bdt['job'].replace(['retired', 'student', 'unemployed', 'unknown'], 'other')
```

```python
# cointing new values
bdt.job.value_counts()
```

```python
# 'poutcome deatils'
bdt.poutcome.value_counts()
```

```python
# combine 'unknown' and 'other' as 'other' isn't really match with either 'success' or 'failure'

bdt['poutcome'] = bdt['poutcome'].replace(['other'], 'unknown')
```

```python
bdt.poutcome.value_counts()
```

```python
# Drop contact as every contact has been contacted, and hence its not a worth feature
bdt.drop('contact', axis=1, inplace= True)
```

```python
bdt.head()
```

```python
# convert categories to binary

bdt['default_cat'] = bdt['default'].map( {'yes':1, 'no':0} )
bdt.drop('default', axis=1, inplace= True)
```

```python
bdt.head()
```

```python
# convert categories to binary

bdt['housing_cat'] = bdt['housing'].map( {'yes':1, 'no':0} )
bdt.drop('housing', axis=1, inplace= True)
```

```python
bdt.head()
```

```python
# convert categories to binary

bdt['loan_cat'] = bdt['loan'].map( {'yes':1, 'no':0} )
bdt.drop('loan', axis=1, inplace=True)
```

```python
bdt.head()
```

```python
# Drop month and day as they don't have any significant meaning
bdt.drop( ['month', 'day'], axis=1, inplace=True)
```

```python
bdt.head()
```

```python
# convert categories to binary

bdt['deposit_cat'] = bdt['deposit'].map( {'yes':1, 'no':0 })
bdt.drop('deposit', axis=1, inplace=True)
```

```python
# pdays: number of days that passed by after the client was last contacted from a previous campaign
#       -1 means client was not previously contacted

print('The no. of customers that have never been contacted: ', len(bdt[ (bdt.pdays == -1)]))
print('Maximum values on pdays: ', bdt.pdays.max())
```

```python
# map pdays =-1 into large value( ex 10000) to indicate that it is so far 
# in the past that it has no effect
bdt.loc[bdt['pdays'] == -1, 'pdays'] = 10000
```

```python
# now convert pdays to recent pdays
# greater the value of a cell in pdays, lesser is the value of it in recent pdays

bdt['recent_pdays'] = np.where(bdt['pdays'], 1/bdt.pdays, 1/bdt.pdays)

# Drop 'pdays'
bdt.drop('pdays', axis=1, inplace= True)
```

```python
bdt.tail()
```

### Standarization and performing PCA

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = bankcl.drop('deposit_cat', 1)
y = bankcl.deposit_cat
```

```python
ss = StandardScaler()
```

```python
sc_dt = ss.fit_transform(x)
```

```python
sc_dt
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(.90)
```

```python
pca_dt = pca.fit_transform(sc_dt)
```

```python
pca_dt
```

```python
pca_dt.shape
```

```python
train_x, test_x, train_y, test_y = train_test_split( pca_dt, y, test_size = 0.2, random_state =50)
```

```python
# Decision tree with depth = 2

dt2 = tree.DecisionTreeClassifier(random_state=1, max_depth=2)
dt2.fit(train_x, train_y)
```

```python
pred_y = dt2.predict(test_x)

print('Training Score : ', dt2.score(train_x, train_y))
print('Testing Score : ', dt2.score(test_x, test_y))
print('Accuracy Score : ', metrics.accuracy_score(test_y, pred_y))
```

```python
# Decision tree: To the full depth
dtf = tree.DecisionTreeClassifier()
dtf.fit(train_x, train_y)
```

```python
pred_yf = dtf.predict(test_x)
```

```python
print("Training score: ",dtf.score(train_x, train_y))
print("Testing score: ",dtf.score(test_x, test_y))
print('Accuracy Score: ',metrics.accuracy_score(test_y, pred_yf))
```
