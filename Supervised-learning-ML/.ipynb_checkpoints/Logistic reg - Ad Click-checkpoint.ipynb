```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#to preprocess the data using StandardScaler =>
import sklearn.preprocessing as prep
#to create a confusion matrix =>
from sklearn.metrics import confusion_matrix
```

```python
#standardscalar = prep.StandardScaler
#load data in 'dt' from csv file
dt = pd.read_csv("C:\\Users\\Admin\\Desktop\\data sets\\Social_Network_Ads.csv")
dt.head(10)
```

```python
#real_x is a 2D array which contans age and estimated salary in each of it's cell 
real_x = dt.iloc[:, [2,3]].values
real_x
```

```python
#real_y is a 1D array which contains the purchased value either 1 or 0
real_y = dt.iloc[:, 4].values
real_y
```

```python
#train_test_split function for real_x and real_y
train_x, test_x, train_y, test_y = train_test_split(real_x, real_y, test_size=0.25, random_state=0)
```

```python
train_x
```

```python
#Standard scaler reduces the gap between age and salary it creates a floating
#numbers 

scaler = prep.StandardScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.fit_transform(test_x)
train_x
```

```python
#create classifier object for LogisticRegression class
classifier_LR = LogisticRegression(random_state = 0)
#train the algorithm
classifier_LR.fit(train_x, train_y)
```

```python
#give the test_x and take the prediction 
y_pred = classifier_LR.predict(test_x)

y_pred
```

```python
test_y
```

```python
#create a confussion matrix to see the probalility of the correctness of the data
####################
#we can use a Confusion Matrix to evaluate exactly how accurate our Logistic Regression model is. 
#This function will compare the calculated results in our y_pred vector to the actual observed results in y_test to
#determine how similar they are. The more values that match, the higher the accuracy of the classifier.
c_m = confusion_matrix(test_y, y_pred)
c_m

#the top-left and bottom-right quadrants will tell us how many predictions were correct - 
#while the other two values indicate how many were in correct. We can divide the summation of those first two numbers by 
#the total number of samples to give us a percentage of accuracy.
```

```python
from matplotlib.colors import ListedColormap
#take the data from train_x and train_y so that our data is available for further use
X_set, y_set = train_x, train_y

#np.meshgrid is creating 2 arrays from np.arange
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1,step =  0.01  ),
                    np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1,step =  0.01  ))

#contourf draws a hypothesis line separating purchased or not-purchased data points
plt.contourf(X1, X2, classifier_LR.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
            alpha = 0.5, cmap = ListedColormap(('red', 'green')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
un_y = np.unique(y_set)
for i, j in enumerate(un_y):
    plt.scatter(X_set[y_set == j,0], X_set[y_set == j, 1],
              c = ListedColormap(('red', 'green'))(i), label =j)
    
plt.title("Logistic Regression(Training set)")
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
#print(X1)
#print(X_set)
#un_y
```
